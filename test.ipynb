```python
import numpy as np
import matplotlib.pyplot as plt

# File path (update if needed)
file_path = "/Users/kanishkkarthick/Documents/Official/final itration /OUTPUT/airfoils/soaring/airfoil_00006_Accipiter_butleri_AR1.35.dat"

# Read file
x = []
y = []

with open(file_path, 'r') as f:
    for line in f:
        # Skip comments and empty lines
        if line.strip().startswith('#') or len(line.strip()) == 0:
            continue
        
        parts = line.split()
        if len(parts) == 2:
            try:
                xi = float(parts[0])
                yi = float(parts[1])
                x.append(xi)
                y.append(yi)
            except:
                continue

# Convert to numpy arrays
x = np.array(x)
y = np.array(y)

# Plot
plt.figure(figsize=(10, 4))
plt.plot(x, y)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Airfoil Shape (AVONET - Strix woodfordii)")
plt.xlabel("x/c")
plt.ylabel("y/c")
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt



x, y = [], []

with open(file_path, 'r') as f:
    for line in f:
        if line.strip().startswith('#') or not line.strip():
            continue
        
        parts = line.split()
        if len(parts) == 2:
            try:
                x.append(float(parts[0]))
                y.append(float(parts[1]))
            except:
                pass

x = np.array(x)
y = np.array(y)

# ---- TRUE SCALE PLOT ----
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=1)

# 🔑 CRITICAL LINE (keeps geometry EXACT)
ax.set_aspect('equal', adjustable='datalim')

# Remove any auto stretching
ax.set_xlim(min(x), max(x))
ax.set_ylim(min(y), max(y))

ax.set_xlabel("x/c")
ax.set_ylabel("y/c")
ax.set_title("Airfoil (True Geometry - No Distortion)")
ax.grid()

plt.show()
```
